```python
# 1/ df1, df2, df3, df4 giống nhau, cách tạo khác nhau 

import pandas as pd

# Tạo Dataframe từ list:
df1= pd.DataFrame(
    [[12000,1560],
    [14500,2350],
    [13200, 1890],
    [15700,1678]],
    index = ["VNM", "TCB", "MSN", "HSG"],
    columns= ['Price','Quantity']
)
print('df1:\n', df1,'\n')

# Tạo Dataframe từ Dictionary of List:
df2 = pd.DataFrame(
    {'Price': [12000, 14500, 13200, 15700],
    'Quantity': [1560, 2350, 1890, 1678]},
    index= ["VNM", "TCB", "MSN", "HSG"]
)
print('df2:\n', df2,'\n')

# Tạo Dataframe từ List of Dicts:
df3 = pd.DataFrame(
    [{'Price':12000,'Quantity':1560},
     {'Price':14500,'Quantity':2350},
     {'Price':13200,'Quantity':1890},
     {'Price':15700,'Quantity':1678}],
     index= ["VNM", "TCB", "MSN", "HSG"]
)
print('df3:\n', df3,'\n')

#  Tạo Data frame từ Dict of Series
df4 = pd.DataFrame(
    {'Price': pd.Series([12000, 14500, 13200, 15700],index=["VNM", "TCB", "MSN", "HSG"]),
     'Quantity':pd.Series([1560, 2350, 1890, 1678],index=["VNM", "TCB", "MSN", "HSG"])}
)
print('df4:\n', df4,'\n')
```

```python
# 7/
import pandas as pd

index = pd.MultiIndex.from_tuples(
    [('Food', 'VNM'),
    ('Bank', 'TCB'),
    ('Food', 'MSN'),
    ('Bank', 'ACB'),
    ('Food', 'VHM'),
    ('Food','CSG'),],
    names=['Group', 'Code'])

columns = pd.MultiIndex.from_tuples(
    [('Bid', 'Price'),
    (' ', 'Quantity'),
    (' ','PE')])
df = pd.DataFrame(
    [(12000,1560, 2.34),
    ( 14500,2350,5.43),
    ( 13200,1890,4.35),
    (15700,1678,3.27),
    ( 16340,890,6.12),
    (13800,1235,4.35 )],
     index= index, columns= columns)


print('df:\n', df,'\n')

print('Xem Sort Index theo Group:\n')
print(df.sort_index(level='Group'))

print('Xem Sort Index theo Code:\n')
print(df.sort_index(level='Code'))

print('Xem Sort Index theo Code, chiều giảm dần:\n')
print(df.sort_index(level='Code',ascending= False))

print('Sort Index theo Group, kết quả ghi lại trong df:\n')
print(df.sort_index(level='Group',inplace=True))
print(df,'\n')

print('Sort Index theo Cột, kết quả ghi lại trong df:\n')
print(df.sort_index(axis=1, inplace= True))
print(df,'\n')

print('Xem Sort theo PE, sau đó theo Bid Price, theo thuật toán quicksort:\n')
sorted_df= df.sort_values(by= [(' ','PE'),('Bid','Price')],kind='quicksort')
print(sorted_df)
```

```python
# 8/
import pandas as pd

df = pd.DataFrame(
        {"Price": [12000, 14500, 13200, 15700, 16340, 13800],
        "Quantity": [1560, 2350, 1890, 1678, 890, 1235], 
        "PE": [2.34, 5.43, 4.56, 3.27, 6.12, 4.35]},
        index= ["VNM", "TCB", "MSN", "ACB", "VHM", "CSG"]
)
print('df:\n', df,'\n')

print('df gồm các cột có tên:\n')
for i in df:
    print(i)
    
print('Giá trị ở mỗi cột như sau:\n')
for key,value in df.iteritems():
    print(key,'\n') 
    print(value,'\n')
    
print('Thông tin của mỗi cổ phiếu như sau:\n')
for row_index,row in df.iterrows():
    print(row_index,'\n')
    print(row,'\n')
    
for row in df.itertuples():
    print(row)
```

```python
# 10/
import pandas as pd
import numpy as np

df = pd.DataFrame([( 12000, 1560, 2.34),
                   (14500, 2350, 5.43),
                   (13200, 1890, 4.56),
                   (15700, 1678,  3.27),
                   (16340, 890, 6.12),
                   (13800, 1235, 4.35 )],
                  index=['VNM', 'TCB', 'MSN', 'ACB', 'VHM', 'CSG'],
                  columns=('Price', 'Quantity','PE'))

print('df:\n', df,'\n')

print('Tính giá trung bình gia quyền (sum(prod(Price*Quantiy))/sum(Quantity)) (Cách 1: chỉ pandas):')
df_multiply = df['Price']*df['Quantity']
cach_1 = df_multiply.sum()/df['Quantity'].sum()
print(f'Cách 1: {cach_1}')

print('Tính giá trung bình gia quyền (sum(prod(Price*Quantiy))/sum(Quantity)) (Cách 2: dùng numpy_ufunc):')
price = list(np.array(df['Price']))
quantity = list(np.array(df['Quantity']))
cach_2 = np.average(price,weights=quantity)
print(f'Cách 2: {cach_2}')
```
