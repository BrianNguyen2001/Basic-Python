```python
x = df_data[['Duration','Pulse','Maxpulse']]
y = df_data['Calories']
x, y = np.array(x), np.array(y)
x = sm.add_constant(x)
```

```python
model = sm.OLS(y,x)
result = model.fit()
print(result.summary())
```

```python
df_income = pd.read_csv(r"C:\Users\ASUS\Desktop\Học\SampleData\income.csv")
df_income
```

```python
y = df_income['Expenditure']
x = df_income['Income']
y,x = np.array(y), np.array(x)
x = sm.add_constant(x)
model_linar = sm.OLS(y,x)
result_linar = model_linar.fit()
```

```python
print(result_linar.summary())
```

```python
x_new_pre = np.array([[12],[30]])
x_new_pre = sm.add_constant(x_new_pre)
predicted = result_linar.predict(x_new_pre)
print(f'Predict Expenditure base on Income:{predicted}')
print(f'Mean: \n{df_income.mean()}')
```

```python
plt.scatter(df_income['Income'],df_income['Expenditure'])
Params = result_linar.params
plt.plot(df_income['Income'],result_linar.params[0]+ result_linar.params[1]*df_income['Income'])
```
